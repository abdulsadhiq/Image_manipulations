```python
#Gray scale effect # B/W
bw_hrs=img_ary.copy()
scl=[0.2989, 0.5870, 0.1140]
sum_px=0
for clr in range(img_ary.shape[2]):
    sum_px+=img_ary[:,:,clr]*scl[clr]
for clr in range(img_ary.shape[2]):
    bw_hrs[:,:,clr]=sum_px
```

```python
#Gray scale effect # B/W - using pixel average
bw_hrs2=img_ary.copy()
for clr in range(img_ary.shape[2]):
    bw_hrs2[:,:,clr]=img_ary.mean(axis=2)
```

```python
# Mirror effect
tot_cl=img_ary.shape[1]-1
hrs_mirr=img_ary.copy()
for cl in range(tot_cl):
    rev_cl=tot_cl-cl
    hrs_mirr[:,rev_cl,:]=img_ary[:,cl,:]
```
